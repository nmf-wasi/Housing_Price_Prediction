### Co-relation Matrix

```python
corr = train_df.corr(numeric_only=True)
corr_sorted=corr["SalePrice"].sort_values(ascending=False)
corr_sorted
```

```python
import matplotlib.pyplot as plt
import os
corr=train_df.select_dtypes(include=["int64", "float64"]).corr()
os.makedirs('../images', exist_ok=True)

plt.figure(figsize=(14, 10))
plt.imshow(corr, cmap="coolwarm", aspect="auto")
plt.colorbar()

plt.xticks(range(len(corr)), corr.columns, rotation=90)
plt.yticks(range(len(corr)), corr.columns)

plt.title("Correlation Matrix")
plt.tight_layout()
plt.savefig("../images/correlation_matrix.png", dpi=300)
plt.show()
```

```python
feature_list = corr_sorted.drop("SalePrice").head(10).index.tolist()
print(feature_list)

for feature in feature_list:
    plt.hist(train_df[feature], bins=30)
    plt.title(f"Distribution of {feature}")
    plt.xlabel(feature)
    plt.ylabel("Number of houses")
    plt.savefig(f"../images/{feature}_distribution.png")
    plt.show()
```

### Categorical Feature Exploration

#### Housing Style

```python
train_df["HouseStyle"].value_counts().plot(
    kind="bar",
    figsize=(7,4)
)

plt.title("House Style Distribution")
plt.xlabel("House Style")
plt.ylabel("Count")
plt.show()
```

### Does this affect Sales Price

```python
train_df.groupby("HouseStyle")["SalePrice"].mean().sort_values()
```

```python
train_df.groupby("HouseStyle")["SalePrice"].mean().sort_values().plot(kind="bar")
```

Difference on house style affects SalePrice.

```python
cat_cols = train_df.select_dtypes(include=["object", "string"]).columns

for col in cat_cols:
    print(f"{col}: {train_df[col].nunique()} unique values")
```

Neighborhood: 25 unique values

Exterior1st: 15 unique values

Exterior2nd: 16 unique values

It's hard to encode all these with one hot encoding. Let's see how much these matter!

```python
train_df.groupby("Neighborhood")["SalePrice"].mean().sort_values()
```

> Neighborhood appears to have a strong relationship with sale price. Houses in different neighborhoods have noticeably different average prices, suggesting that location is an important predictor.

```python
train_df.groupby("Exterior1st")["SalePrice"].mean().sort_values()
```

> The primary exterior material also shows differences in average sale price. This suggests that exterior material may contribute to predicting house prices, although further analysis is needed to determine its importance alongside other features.

```python
train_df.groupby("Exterior2nd")["SalePrice"].mean().sort_values()
```
